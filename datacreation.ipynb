```python
import pandas as pd 
weather_df=pd.read_csv("weather_info.csv")

weather_df.rename(columns={'YEAR': 'year', 'MO': 'month', 'DY': 'day'}, inplace=True)
weather_df.head()
```

```python
weather_df['Date'] = pd.to_datetime(weather_df[['year', 'month', 'day']])
```

```python
weather_df.head()
```

```python
weather_df.drop(columns=['year', 'month', 'day'], axis=1, inplace=True)
```

```python
weather_df.head()
```

```python
first_column=weather_df.pop('Date')
weather_df.insert(0, 'Date', first_column)
```

```python
weather_df.rename(columns={'T2M':'Temp', 'RH2M':'Relative Humidity'}, inplace=True)
weather_df.head()
```

```python
demand_df=pd.read_csv("daily_demand_data.csv")
demand_df['Date'] = pd.to_datetime(demand_df['Date'], format='%m/%d/%Y')
demand_df.head()
```

```python
final_df = pd.merge(demand_df, weather_df, on='Date', how='inner')
```

```python
final_df.head()
```

```python
final_df.drop(columns=['Temp', 'Relative Humidity_y'], inplace=True)
```

```python
final_df.head()
```

```python
final_df.rename(columns={'Relative Humidity_x': 'Relative Humidity'}, inplace=True)
final_df.head()
```

```python
final_df.to_csv('final_dataset.csv', index=False)
```
